```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("base.csv", sep=";")
```

```python
data.head()
```

```python
data.shape
```

```python
# Qual o valor de correlação entre as variáveis qtd_pessoas e valor da diária? 
data['qtd_pessoas'].corr(data['valor_diaria'])
```

```python
#CRIAR O MODELO DE REGRESSÃO LINEAR
from sklearn.linear_model import LinearRegression

linear_regg = LinearRegression()
```

```python
#PREDITOR
X = data.iloc[:,0].values 
#Transformando em coluna
X = X.reshape(-1,1)
#TARGET
y = data.iloc[:,1].values
```

```python
X.shape
```

```python
#TREINANDO O MODELO
linear_regg.fit(X,y)
```

```python
# Qual o valor retornado intercept da previsão? 
linear_regg.intercept_
```

```python
previsao = linear_regg.predict(X)
```

```python
#  Qual o valor da previsão realizada pelo modelo de regressão para a qtd_pessoas = 20 ? 
linear_regg.intercept_ + linear_regg.coef_ * 20
```

```python
#  Qual o valor da previsão realizada pelo modelo de regressão para a qtd_pessoas = 57 ? 
linear_regg.intercept_ + linear_regg.coef_ * 57
```

```python
import plotly.express as px 
import seaborn as sns
```

```python
 #Qual a difença entre o valor predito e o valor real do valor da diária para a quantidade de pessoas = 25 
grafico = px.scatter(x = X.ravel(), y = y)
grafico.add_scatter(x = X.ravel(), y = previsao, name = "Previsões")
grafico.show()
```

```python
# Qual foi o score obtido pelo algoritmo de regressão? 
linear_regg.score(X,y)
```

```python
# Qual o valor obtido para mean_absolute_error? 
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y, previsao)
```
