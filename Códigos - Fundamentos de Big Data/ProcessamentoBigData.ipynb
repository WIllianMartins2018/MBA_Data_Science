# **ALTERANDO A DISPOSIÇÃO DAS COLUNAS**

```python
base_completa[['cod_cliente', 'genero', 'idade', 'peso', 'colesterol', 'id_estado', 'estado', 'sigla_estado', 'pais']]
```

```python
base_completa = base_completa[['cod_cliente', 'genero', 'idade', 'peso', 'colesterol', 'id_estado', 'estado', 'sigla_estado', 'pais']].reset_index(drop=True)
```

```python
base_completa
```

# **ANÁLISE DE DADOS**

```python
from sklearn.cluster import KMeans
```

```python
dados_cliente = base_completa[['peso', 'colesterol']]
kmeans = KMeans(n_clusters=4)
kmeans.fit(dados_cliente)
dados_cliente
```

```python
kmeans.inertia_
```

```python
base_completa['cluster'] = kmeans.fit_predict(dados_cliente)
```

```python
base_completa
```

```python
centroides = kmeans.cluster_centers_
centroides
```

```python
base_completa.groupby(['cluster']).mean()
```

```python
base_completa.groupby(['cluster']).max()
```
