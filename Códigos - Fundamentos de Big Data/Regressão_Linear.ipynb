**PREVER O CUSTO DA DIÁRIA DE UM IMÓVEL BASEADO AO NÚMERO DE PESSOAS**

```python
import pandas as pd
import plotly.express as px 
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv("base_custo_diaria.csv", sep=";")
```

```python
data.head()
```

```python
#Algumas estatísticas básicas da base
data.describe()
```

```python
#Buscamos todos os registros da primeira coluna e convertemos em array
X_custo_diaria = data.iloc[:,0].values
X_custo_diaria
```

```python
#Criação da variavel que tem que ser predita
y_custo_diaria = data.iloc[:,1].values
y_custo_diaria
```

```python
#Criação de uma matriz de correlação
np.corrcoef(X_custo_diaria, y_custo_diaria)
```

```python
corr = data.corr()
plot = sns.heatmap(corr, annot= True, linewidths=.3)
plot
```

```python
#TRATANDO OS DADOS PARA O MODELO
X_custo_diaria.shape
```

```python
#O resultado mostra que temos um array com apenas uma coluna. No entando, precisamos ter uma matriz com 2 colunas
X_custo_diaria = X_custo_diaria.reshape(-1,1)
X_custo_diaria
```

```python
X_custo_diaria.shape
```

```python
#Criando o Modelo de Regressão Linear
from sklearn.linear_model import LinearRegression

linear_regg = LinearRegression()
```

```python
type(linear_regg)
```

```python
#REALIZANDO O TREINAMENTO DO MODELO
linear_regg.fit(X_custo_diaria, y_custo_diaria)
```

```python
#Verificando a intercepção
linear_regg.intercept_
```

```python
#Verificando o coeficiente angular da reta
linear_regg.coef_
```

```python
#REALIZANDO PREVISÕES
previsao_precos = linear_regg.predict(X_custo_diaria)
```

```python
#Visualizando as previsões
previsao_precos
```

```python
#VISUALIZANDO NO GRÁFICO
#Transformando em vetor
X_custo_diaria.ravel()
```

```python
grafico = px.scatter(x = X_custo_diaria.ravel(), y = y_custo_diaria)
grafico.show()
```

```python
grafico = px.scatter(x = X_custo_diaria.ravel(), y = y_custo_diaria)
grafico.add_scatter(x = X_custo_diaria.ravel(), y = previsao_precos, name = "Previsões")
grafico.show()
```

```python
#Verificando os valores de uma determinada previsão baseada na qtd de pessoas
qtd_pessoas = 4
linear_regg.intercept_ + linear_regg.coef_ * qtd_pessoas
```

```python
#VERIFICANDO A QUALIDADE DO ALGORITMO. QUANTO MAIS PROXIMO DE 1 MELHOR
linear_regg.score(X_custo_diaria, y_custo_diaria)
```

```python
#VISUALIZANDO OS ERROS DO MODELO
from sklearn.metrics import *
MSE = mean_squared_error(y_custo_diaria, previsao_precos)
```

```python
MSE
```

```python
#CALCULANDO O MAE
mean_absolute_error(y_custo_diaria, previsao_precos)
```

```python
#CALCULANDO O RMSE
np.sqrt(MSE)
```
