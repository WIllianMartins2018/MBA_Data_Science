```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataframe = pd.read_csv("country_vaccinations.csv", sep=",", encoding='latin-1')
```

```python
dataframe.head()
```

```python
#Verificando os dados referentes ao Brasil por dia
dataframe.query('country == "Brazil"').sort_values('date', ascending=False)[:2]
```

```python
#Verificando o numero de pa√≠ses distintos
dataframe['country'].nunique()
```

```python
#Identificando a data mais recente de coleta
ultima_data = dataframe['date'].max()
ultima_data
```

```python
#filtrando a ultima data
ultima_data = dataframe[dataframe.date == ultima_data]
ultima_data
```

```python
ultima_data.shape
```

```python
#Ultima data em que teve vacina√ß√£o em todos os pa√≠ses
dataframe.groupby('country')['date'].max().sort_values(ascending=False)
```

```python
#Criar um top 10 pa√≠ses com maior n√∫mero de vacinados
imunizados_por_pa√≠s = dataframe.groupby('country')['total_vaccinations'].max().sort_values(ascending=False)
imunizados_por_pa√≠s
```

```python
type(imunizados_por_pa√≠s)
```

```python
imunizados = imunizados_por_pa√≠s.to_frame().reset_index()
imunizados
```

```python
#Determinar a vacina mais utilizada
vacina_por_pais = dataframe[['country', 'date', 'vaccines']]
vacina_por_pais
```

```python
vacina_por_pais.groupby('vaccines')['country'].nunique().sort_values(ascending=False)
```

```python
vacinas_pais = dataframe.groupby('country')['vaccines'].unique()
vacinas_pais = vacinas_pais.to_frame().reset_index()
```

```python
vacinas_pais
```

```python
#Qual pa√≠s possui o maior n√∫mero de vacinados por dia
dataframe.groupby('country')['daily_vaccinations'].max().sort_values(ascending=False)
```

```python
#Qual a m√©dia de vacina√ß√£o por dia em cada pa√≠s
media_pais = dataframe.groupby('country')['daily_vaccinations'].mean().sort_values(ascending=False)
media_pais = media_pais.to_frame().reset_index()
media_pais
```

```python
#Renomeando as Colunas
media_pais   = media_pais.rename(columns={'country': 'pais', 'daily_vaccinations': 'media_vacinacao'})
vacinas_pais = vacinas_pais.rename(columns={'country':'pais', 'vaccines':'vacina_utilizada'})
imunizados   = imunizados.rename(columns={'country':'pais', 'total_vaccinations': 'Imunizados'}) 
```

```python
dados_tratados = pd.merge(imunizados, media_pais, how='inner')
```

```python
dados_tratados = pd.merge(dados_tratados, vacinas_pais)
dados_tratados
```

```python
#Exportando os dados
dados_tratados.to_csv('estastistica_pais.csv', index= True, sep=';')
dados_tratados.to_json('estastistica.json', orient='records')
```
