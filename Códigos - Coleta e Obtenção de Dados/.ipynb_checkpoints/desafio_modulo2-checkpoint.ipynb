```python
ds_jogadores = "dados_jogadores.csv"
ds_chess = "lista_jogadores_chess.json"
```

```python
local_arquivo = os.path.join(DIRETORIO,ds_jogadores)
local_arquivo
```

```python
df_jogadores = pd.read_csv(local_arquivo, sep=';', encoding='latin-1')
```

```python
df_jogadores.tail()
```

```python
## avaliação de campos nulos
df_jogadores.isnull().sum()
```

```python
## ajustando os nulos
df_jogadores.loc[df_jogadores['num_vitorias'].isnull(), 'num_vitorias'] = df_jogadores['total_partidas']-df_jogadores['num_derrotas'] 
df_jogadores.loc[df_jogadores['num_derrotas'].isnull(), 'num_derrotas'] = df_jogadores['total_partidas']-df_jogadores['num_vitorias'] 
df_jogadores.loc[df_jogadores['total_partidas'].isnull(), 'total_partidas'] = df_jogadores['num_vitorias']+df_jogadores['num_derrotas'] 
```

```python
## tratando o codigo do jogador
df_jogadores.jogador = df_jogadores.jogador.str.split(' ').str[1]
```

```python
engine = sqlalchemy.create_engine(string_conexao)
conn = engine.connect()
```

```python
paises = list(df_jogadores.pais.unique())
```

```python
paises
```

```python
jogos = list(df_jogadores.jogo.unique())
```

```python
jogos
```

```python
df_jogadores.shape 
```

```python
for pais in paises:
    try:
        query = f"""insert into pais (nome_pais) 
                     values ('{pais}')"""
        conn.execute(query)
        print(f"Registro inserido com sucesso: {pais}")
    except Exception as e:
        print(f"Não foi possível inserir o regristro {pais}. O erro foi encontrado foi: {e}")   
```

```python
for jogo in jogos:
    try:
        query = f"""insert into jogo (nome_jogo) 
                     values ('{jogo}')"""
        conn.execute(query)
        print(f"Registro inserido com sucesso: {jogo}")
    except Exception as e:
        print(f"Não foi possível inserir o regristro {jogo}. O erro foi encontrado foi: {e}")   
```

```python
df_jogos = pd.read_sql('jogo', con=conn)
df_paises = pd.read_sql('pais', con=conn)
```

```python
dict_jogos = df_jogos.set_index(df_jogos.nome_jogo)['cod_jogo'].to_dict()
dict_jogos
```

```python
dict_paises = df_paises.set_index(df_paises.nome_pais)['cod_pais'].to_dict()
dict_paises
```

```python
df_jogadores.pais = df_jogadores.pais.map(dict_paises)
df_jogadores
```

```python
df_jogadores.jogo = df_jogadores.jogo.map(dict_jogos)
df_jogadores
```

```python
for jogador in df_jogadores.itertuples():
    cod_jogador = jogador.jogador
    genero = jogador.genero
    data_nascimento =  datetime.datetime.strptime(jogador.data_nascimento, "%d/%m/%Y").strftime('%Y-%m-%d')
    num_vitorias = jogador.num_vitorias
    num_derrotas = jogador.num_derrotas
    total_partidas = jogador.total_partidas
    cod_pais = jogador.pais
    cod_jogo = jogador.jogo
    try:        
        query = f"""insert into jogador (cod_jogador, genero, data_nascimento, num_vitorias, num_derrotas, total_partidas, cod_pais, cod_jogo) 
                     values ('{cod_jogador}','{genero}', '{data_nascimento}', '{num_vitorias}', '{num_derrotas}', '{total_partidas}', '{cod_pais}', '{cod_jogo}')"""
        print(query)
        conn.execute(query)
        print(f"Registro inserido com sucesso: {cod_jogador}")
    except Exception as e:
        print(f'Não foi possível inserir o registro: {cod_jogador} Erro: {e}')
    
```
